По желанию вы можете исключить ещё какие-то признаки, или, наоборот, сформировать новые на основе имеющихся. Цель: реализовать функцию, которая будет принимать на вход набор признаков для объекта выборки и возвращать прогноз целевой переменной (0 или 1). Например, вы можете просто сложить все признаки и возращать ноль, если сумма отрицательная, а в противном случае единицу. Но наверняка можно придумать лучше!
Данные уже разделены на набор для обучения, который вы можете использовать как угодно, и тестовый, на котором нужно будет проверить вашу точность. Постарайтесь достичь точности хотя бы 0.6.

```python
# можем вручную получить мнк оценку по формуле (X^tX)^-1X^tY потому что матрица маленькая
# признаки отмасштабированы
# выбросы только не пофикшены
def evalB(X_train: np.array, y_train:np.array) -> np.array:
    return np.linalg.inv((X_train.T.dot(X_train))).dot(X_train.T).dot(y_train)

def predict(X_test: np.array) -> np.array:
    return list(map(lambda x: 1 if x >=0.1 else 0, X_test.dot(coef)))

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

X = data.drop('credit_risk', 1).values
y = data['credit_risk'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
std_scaler = StandardScaler().fit(X_train)

X_train = std_scaler.transform(X_train)
X_test = std_scaler.transform(X_test)

coef = evalB(X_train, y_train)

accuracy_score(y_test, predict(X_test))
```
